### Boxplot dos atributos:

#### Pacotes

```python
% matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
import sys
import numpy as np
plt.rcParams["figure.figsize"] = (25,16)
```

#### Definição da função de plot

```python
def plot_attributes(att):
    data_att = pd.read_json('data/output_attributes.json')
    
    data_plot = pd.DataFrame(data_att[att])
    data_plot.plot(kind='box', title=att)
```

#### Kills: É a quantidade de heróis que um jogador mata em uma partida

```python
plot_attributes('kills')
```

#### Deaths: É a quantidade de vezes que um jogador morre em uma partida

```python
plot_attributes('deaths')
```

#### Assistências: É a quantidade de vezes que o jogador ajudou um aliado a matar um inimigo

```python
plot_attributes('assists')
```

#### Denies: Quantidade de vezes que o jogador conseguiu negar uma unidade aliada

```python
plot_attributes('denies')
```

#### Gold per minute: Quantidade de ouro o jogador conseguiu no jogo por minuto

```python
plot_attributes('gpm')
```

#### Hero damage: Quantidade de dano que o jogador deu em heróis inimigos

```python
plot_attributes('hero_damage')
```

#### Hero healing: Quantidade de cura que o jogador efetuou no jogo

```python
plot_attributes('hero_healing')
```

#### Last hits: Quantidades de unidades inimigas que o jogador finalizou

```python
plot_attributes('lh')
```

#### XP per minute: Quanto de experiência o jogador obteve por minuto

```python
plot_attributes('xp_p_min')
```
