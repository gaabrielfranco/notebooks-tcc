##### Um exemplo de clusterização

```python
import numpy as np
from sklearn.cluster import KMeans
```

- Normalização dos atributos

```python
def normalizes(x):
    x_norm = []
    minimum = np.min(x)
    maximum = np.max(x)
    for i in x:
        x_norm.append((i - minimum) / (maximum - minimum))

    return x_norm, minimum, maximum
```

- Volta os atributos para não normalizados

```python
def un_normalizes(m, minimum, maximum):
    x_un_norm = []
    for i in m:
        x_un_norm.append(i * (maximum - minimum) + minimum)

    return x_un_norm
```

- Printa o sumário da classificação

```python
def summary(eff, tabs=0):
    if tabs == 0:
        print('Media: ' + str(np.average(eff)) + '\n')
        print('Desvio: ' + str(np.std(eff)) + '\n')
        print('Coef. Variacao: ' +
              str(np.std(eff) / np.average(eff)) + '\n')
        print('Minimo: ' + str(np.min(eff)) + '\n')
        print('Maximo: ' + str(np.max(eff)) + '\n')
    else:
        for i in range(len(eff)):
            print('\t\tCluster %d:\n' % (i + 1))
            print('\t\t\tMedia: ' + str(np.average(eff[i])) + '\n')
            print('\t\t\tDesvio: ' + str(np.std(eff[i])) + '\n')
            print('\t\t\tCoef. Variacao: ' +
                  str(np.std(eff[i]) / np.average(eff[i])) + '\n')
            print('\t\t\tMinimo: ' + str(np.min(eff[i])) + '\n')
            print('\t\t\tMaximo: ' + str(np.max(eff[i])) + '\n')
```

- Função que faz a classificação

```python
def classification(k, data):
    data_norm, min_norm, max_norm = normalizes(data)

    inertia = []
    cluster_centers_eff = []
    eff_f = []
    i = 0

    for j in range(k):
        eff_f.append([])

    print('Execucao sem podas de outliers com dados normalizados por n-partidas e min-max\n\n')

    print('Execucao com semente ' + str(i) + ':\n')
    km = KMeans(n_clusters=k, random_state=i)
    labels = km.fit_predict(data_norm)

    inertia.append(km.inertia_)
    print('\tInertia da execucao com semente ' + str(i) + ':\n')
    print('\t' + str(km.inertia_) + '\n\n')
    count = [0] * k
    for l in labels:
        count[l] += 1
    print('\tContagem da execucao com semente ' + str(i) + ':\n')
    print('\t' + str(count) + '\n\n')

    centroids = un_normalizes(
        km.cluster_centers_[:], min_norm, max_norm)
    print(
        '\tCentroides da execucao com semente ' + str(i) + ':\n')
    for j in centroids:
        print('\t' + str(j) + '\n')
    print('\n')

    for j in centroids:
        cluster_centers_eff.append(
            j[0] - j[1] + j[2] + max(j[3], j[4]))

    for j, d in enumerate(data_norm):
        eff = d[0] - d[1] + d[2] + max(d[3], d[4])

        eff_f[labels[j]].append(
            abs(cluster_centers_eff[labels[j]] - eff))

    print('\tDados da metrica F: \n')
    summary(eff_f, 1)
    print('\n')

    print('Dados da metrica inertia: \n')
    summary(inertia)
```

- Lê os dados e classifica

```python
data_all = []

fp = open('data/attributes.txt', 'r')

for l in fp:
    parts = l.strip().split()
    for i, p in enumerate(parts):
        parts[i] = int(p)
    if parts[4] >= 5:
        data_all.append(
            list(np.array(parts[1:4] + [parts[5]] + [parts[9]]) / parts[4]))

fp.close()

classification(3, data_all)
```

##### Clusterização por correlação

- Utilizamos a correlação de Pearson dada por:
$$\rho = \frac{\sum_{i=1}^{n} (x_i - \bar x)(y_i - \bar y) }{\sqrt{\sum_{i=1}^{n} (x_i - \bar x) ^ 2}.{\sqrt{\sum_{i=1}^{n} (y_i - \bar y) ^ 2}}} $$ 
onde $x_1, x_2, ..., x_n$ e $y_1, y_2, ..., y_n$ são os valores medidos de ambas as variáveis

- Para cada atributo da base, clusterizamos ele com os quatros atributos que obtiveram o menor módulo da correlação com este atributo

- A correlação dos atributos foi a seguinte:

```python
def correlation_analysis(data, attributes):
    output_data = {}
    for i, attr in enumerate(attributes):
        output_data[attr] = []
        for d in data:
            output_data[attr].append(float(d[i]))

    df = pd.DataFrame(output_data)
    df = df.corr()
    
    return df
    
def read_data():
    data = []
    fp = open('data/attributes.txt', 'r')
    for l in fp:
        parts = l.strip().split()
        for i, p in enumerate(parts):
            parts[i] = int(p)
        if parts[4] >= 5:
            data.append(list(np.array(parts[1:4] + parts[5:]) / parts[4]))
    fp.close()
    return data

def highlight_attributes(att):
    att = att.apply(lambda x: abs(x))
    sorted_att = att.sort_values(ascending=True)
    sorted_list = list(sorted_att[:4])
    return ['background-color: grey' if value in sorted_list else '' for value in att]

attributes_names = ["kills", "deaths", "assists", "denies", "gpm", "hd", "hh", "lh", "xpm"]

data = read_data()

df = correlation_analysis(data, attributes_names)

final_table = df.style.apply(highlight_attributes)

final_table

```

### Análise do parâmetro $k$ no K-Means

- Antes de rodar novamente os experimentos, fiz uma análise de qual é o parâmetro $k$ ótimo para ser utilizado no k-means, já que a escolha de $k = [3, 4, 5]$ foi intuitiva 
- Para fazer essa análise, rodamos o k-means com $k = [3, 100]$ para cada experimento e plotamos o gráfico da inércia por $k$
- Nessa curva, o ponto que a reta tangente a curva tem o ângulo $\theta = 45°$ é o ponto ótimo para o $k$
- A inércia é soma das distâncias quadradas das amostras para o centróide mais próximo e é dada por: 
$$I = \sum_{i}(d(i,cr))$$
onde $cr$ é o centróide mais próximo e $d$ é a distância euclidiana ao quadrado

- ##### Plot da inertia x $k$

```python
%matplotlib inline  
import json
import matplotlib.pyplot as plt

# Fig params
plt.rcParams["figure.figsize"] = (25, 16)
plt.rcParams['font.size'] = 18.0

# Colocar os dados de todos os experimentos aqui
with open('data/output_k_analysis.json') as fh:
    k_analysis = json.load(fh)
    
legends = []

for experiment in k_analysis:
    if experiment != 'all' and experiment != 'kda':
        plt.plot(k_analysis[experiment]['n_clusters'], k_analysis[experiment]['inertia'])
        legends.append(experiment)
    
plt.title("Inertia x K")
plt.ylabel("Inertia")
plt.xlabel("k")
plt.legend(legends)
```

### Outiliers

- Decidimos que vamos podar todos os outliers a partir de agora.
- Para podar, utilizamos o IQR. Se o jogador estiver abaixo do limite inferior ou acima do limite superior em pelo menos um atributo, ele é considerado um outlier

```python
Image(filename='data/Elements_of_a_boxplot_pt.png')
```

Fonte: [Wikipédia](https://pt.wikipedia.org/wiki/Amplitude_interquartil). Acesso em 04/04/2018

- As fórmulas para calcular o IQR e os limites são:
$$IQR = Q3 - Q1$$

$$limite\_inferior = Q1 - 1,5 * IQR$$

$$limite\_superior = Q3 + 1,5 * IQR$$
